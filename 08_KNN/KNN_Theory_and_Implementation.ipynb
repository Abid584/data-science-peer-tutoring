# K-Nearest Neighbors (KNN) - Theory and Implementation


K-Nearest Neighbors (KNN) is a **supervised machine learning algorithm** used for **classification and regression** problems.

### Theory:
- **Lazy Learner**: KNN doesn't learn a model in training. It stores the training dataset.
- **Distance Metric**: Typically uses Euclidean distance to find the closest neighbors.
- **k value**: The number of nearest neighbors used to make predictions.
- **Majority Voting (for classification)**: Class with the most neighbors among the k nearest is selected.
- **Averaging (for regression)**: Average of the values of the k nearest neighbors.

### When to use:
- When you want a simple model that does not assume any distribution.
- Works best with smaller datasets with fewer dimensions.

### Pros:
- Simple and easy to understand.
- No training phase.

### Cons:
- Slow prediction on large datasets.
- Sensitive to the scale of the data.
- Choice of k and distance metric can significantly impact performance.
    

```python
# Importing libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Fitting KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
# Predictions and Evaluation
y_pred = knn.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```


### Observations:
- You can experiment with different values of `k` and observe how accuracy changes.
- Feature scaling is crucial for distance-based algorithms like KNN.
